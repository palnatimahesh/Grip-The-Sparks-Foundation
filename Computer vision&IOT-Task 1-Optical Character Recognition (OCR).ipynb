Character Detector which extracts printed or handwritten words from an image or video.

```python
#Importing Required Libraries
import keras_ocr as kr
import matplotlib.pyplot as plt
```

```python
# Automatically downloads pretrained weights for the detector and recognizer
pipeline=kr.pipeline.Pipeline()
```

```python
#importing images
image_ocr=[
           kr.tools.read(url) for url in["https://hips.hearstapps.com/hmg-prod.s3.amazonaws.com/images/run-motivational-gallery-quote-1-1614804107.jpg?crop=1xw:1xh;center,top&resize=980:*",
                                         "https://hips.hearstapps.com/hmg-prod.s3.amazonaws.com/images/run-motivational-gallery-quote-4-1614888150.jpg?crop=1xw:1xh;center,top&resize=980:*"
                            
                                         ]
]
```

```python
#number of images
len(image_ocr)
```

```python
#Displaying the image
plt.figure(figsize=(10,10))
plt.imshow(image_ocr[0])
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image_ocr[1])
```

```python
prediction_groups=pipeline.recognize(image_ocr)
```

```python
#Get the Predictions
fig, axs = plt.subplots(nrows=len(image_ocr), figsize=(15, 15))
for ax, image, predictions in zip(axs, image_ocr, prediction_groups):
    kr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
